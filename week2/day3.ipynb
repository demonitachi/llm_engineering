```python
def chat(message, history):
    # Helper to convert Gradio's complex content into a simple string
    def format_content(content):
        if isinstance(content, str):
            return content
        if isinstance(content, list):
            # Extract text from [{'text': '...', 'type': 'text'}, ...]
            return " ".join([item["text"] for item in content if "text" in item])
        return str(content)

    # 1. Initialize messages with a system prompt
    messages = [{"role": "system", "content": system_message}]
    
    # 2. Add history (flattening the content for each turn)
    for turn in history:
        messages.append({
            "role": turn["role"],
            "content": format_content(turn["content"])
        })
    
    # 3. Add the current message
    messages.append({
        "role": "user", 
        "content": format_content(message)
    })

    try:
        stream = openrouter.chat.completions.create(
            model="liquid/lfm-2.5-1.2b-instruct:free",
            messages=messages,
            stream=True
        )
        response = ""
        for chunk in stream:
            response += chunk.choices[0].delta.content or ''
            yield response
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
gr.ChatInterface(fn=chat).launch()
```

```python
def chat(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    stream = gemini.chat.completions.create(model=model, messages=messages, stream=True)
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
gr.ChatInterface(fn=chat).launch()
```
