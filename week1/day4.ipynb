```python
import tiktoken
```

```python
encoding = tiktoken.encoding_for_model("gpt-4o")
```

```python
token = encoding.encode("My Name is Kunal Anand")
```

```python
token
```

```python
for i in token:
    token_str = encoding.decode([i])
    print(f"{i}: {token_str}")
```

```python
from openai import OpenAI
```

```python
openai = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
```

```python
messages = [{"role": "system", "content": "You are a helpful assistant."},{"role": "user", "content": "What is the capital of France?"}]
```

```python
response = openai.chat.completions.create(model="gemma3:4b",messages=[{"role": "system", "content": "You are a helpful assistant."},{"role": "user", "content": "What is the capital of France?"}])
```

```python
response.choices[0].message.content
```
